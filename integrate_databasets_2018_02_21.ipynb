### Make new SIDER database : unique pubchem ID, frequency, medDRA term
- Make new df for sider with unique pairs of puchem id, frequency, and meddra term 

```python
# input: dataframe only with columns to be used to make unique identifier
# output: list of unique identifiers for rows of df, based on all columns of df 
# make sure to only input subsetted df with only columns want to be used to create identifier
def make_identifier(df):
    str_id = df.apply(lambda x: '_'.join(map(str,x)), axis = 1)
    return pd.factorize(str_id)[0]

meddra_freq['unique_id'] = make_identifier(meddra_freq[['stitch_id_stereochem', 'frequency', 'umls_id_meddra']])
```

```python
# drop duplicates from sider df based on unique identifier made above
# there are duplicates as a drug can be listed with different meddra_concept_types and meddra_ids that all fall under the same umls_identifier 
sider_uniq_pubchem_freq_umlsID = meddra_freq.drop_duplicates(subset='unique_id', keep="first")
```

```python
# create dictionaries 

# maps drugID to status (1:1 relationship)
drug_status_dict = (drugs.drop(['Name', 'Type', 'SMILES'], axis = 1)).set_index('DrugID')['Status'].to_dict()

# maps targetID to uniprot id (1:1 relationship)
target_uniprot_dict = (targets.drop(['HGNC', 'Species', 'Name'], axis = 1)).set_index('TargetID')['UniProt'].to_dict()

# use for loop to subset df for each unique drug ID and then for that subset, grab targets
# maps drugid to list of targets (targetID), value is a list of targets
drug_target_dict = {} 
drug_ids = list(interactions.DrugID.unique())
for drugID in drug_ids:
    subset = (interactions.drop(['Action'], axis = 1)).loc[(interactions.drop(['Action'], axis = 1))['DrugID'] == drugID]
    target_list = list(subset.TargetID)
    drug_target_dict[drugID] = target_list

# maps drugID to list of names 
drug_name_dict = {} 
drug_ids = list(names.DrugID.unique())
for drugID in drug_ids:
    subset = (names.drop(['NameType'], axis = 1)).loc[(names.drop(['NameType'], axis = 1))['DrugID'] == drugID]
    #subset = names.loc[(names.drop(['NameType'], axis = 1))['DrugID'] == drugID]
    name_list = list(subset.drugname)
    drug_name_dict[drugID] = name_list
```

```python
## make function to convert targetID list to uniprot ID list 
# this function takes in a list of lists of targetIDs and returns the same list but with uniprot IDs 
# some drugs have multiple targets and will have a list of targets 
def targetIDs_to_uniprotIDs(target_list):
    uniprot_final_list = []
    for drug_targets in target_list: 
        uniprot_list_indiv_drug = []
        for indiv_target in drug_targets:
            uniprot_list_indiv_drug.append(target_uniprot_dict[indiv_target])
        uniprot_final_list.append(uniprot_list_indiv_drug)
    return uniprot_final_list
```
